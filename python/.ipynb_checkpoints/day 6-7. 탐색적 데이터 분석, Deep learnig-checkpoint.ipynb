```python
temp = []
for i in df['sex']:
    if i =='female':
        temp.append(1)
    elif i == 'male':
        temp.append(0)
    else:
        temp.append(np.nan)
df['sex'] = temp
df
```

```python

df.info()
```

```python
#혈액형의 A,B,O,AB는 1,2,3,4로 바꾸면 우선순위가 생겨서 문제가 생김. 리스트로 [1000],[0100]...으로 바꾸기

df['survived'] = df['survived'].astype('float')
df['pclass'] = df['pclass'].astype('float')
df['sex'] = df['sex'].astype('float')
df['sibsp'] = df['sibsp'].astype('float')
df['parch'] = df['parch'].astype('float')
df['fare'] = df['fare'].astype('float')
```

```python
df=df[df['age'].notnull()]
df=df[df['fare'].notnull()]
```

```python
#pclass: 0, sex:3 ...
x_data=df.values[:,[0,3,4,5,8]]
```

```python
#survived: 1
y_data=df.values[:,[1]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x_data, y_data, test_size=0.1,random_state=7)
```

```python
X_train.shape,X_test.shape, y_train.shape,y_test.shape
```

```python
import tensorflow as tf
```

```python
tf_X_train = tf.convert_to_tensor(X_train, dtype.flot32)
tf_X_test = tf.convert_to_tensor(X_test, dtype=tf.flot32)
tf_y_train = tf.convert_to_tensor(y_train, dtype=tf.flot32)
tf_y_test = tf.convert_to_tensor(y_test, dtype=tf.flot32)
```

```python
#from tk.keras import Sequential #tf를 못찾음
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

```

```python
model = Sequential()
model.add(Dense(256, input_shape=(6,),activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='mse',optimizer="adam",metircs=['accuracy'])
model.summary()

```

```python
%%time
hist = model.fit(tf_X_train, tf_y_train,validation_data=(tf_X_test,tf_y_test), epochs=500)
```

```python
plt.figure(figsize=(12,8))
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt,show()
```

```python
#디카프리오 3등실, 남자:0, 나이 19, 형제:0, 부모:0, 운임:5

dicaprio_arr = np.array([3.,0.,19.,0.,0.,5.]).reshape(1,6)
dicaprio = tf.convert_to_tensor(dicaprio_arr, dtype=tf.float32)
model.predict(dicaprio)
#실행결과 생존확률 19.1%
```
